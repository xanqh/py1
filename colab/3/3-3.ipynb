ï¼ˆ`return` ãŒå®Ÿè¡Œã•ã‚Œãšã«é–¢æ•°ã®æœ€å¾Œã¾ã§æ¥ãŸã¨ãã‚‚åŒæ§˜ã§ã™ã€‚ï¼‰
`return` ã‚’å¼ãªã—ã§å®Ÿè¡Œã™ã‚‹ã“ã¨ã§ã€é–¢æ•°ã®å‡¦ç†ã‚’é€”ä¸­ã§æŠœã‘ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚

````python
#å…¥åŠ›ã®å¹³å‡ã‚’è¨ˆç®—ã—ã¦è¿”ã™é–¢æ•°average
def average(nums):
    #çµ„ã¿è¾¼ã¿é–¢æ•°ã®sum()ã¨len()ã‚’åˆ©ç”¨
    return sum(nums)/len(nums)

#é–¢æ•°averageã«æ•°å­—ã®ãƒªã‚¹ãƒˆã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
average([1,3,5,7,9])
````

## è¤‡æ•°ã®å¼•æ•°
é–¢æ•°ã¯ä»»æ„ã®æ•°ã®å¼•æ•°ã‚’å—ã‘å–ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚
è¤‡æ•°ã®å¼•æ•°ã‚’å—ã‘å–ã‚‹å ´åˆã¯ã€å¼•æ•°ã‚’ã‚³ãƒ³ãƒã§åŒºåˆ‡ã‚Šã¾ã™ã€‚
ã“ã‚Œã‚‰ã®å¼•æ•°åã¯é‡è¤‡ã—ãªã„ã‚ˆã†ã«ã—ã¾ã—ã‚‡ã†ã€‚

````python
#3ã¤ã®å¼•æ•°ãã‚Œãã‚Œã«æ¸¡ã•ã‚ŒãŸå€¤ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting(en, fr, de):
    print(en + ', ' + fr + ', ' + de)

#é–¢æ•°greetingã«3ã¤ã®å¼•æ•°ã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting('Hello', 'Bonjour', 'Guten Tag')
````

é–¢æ•°ã¯ç•°ãªã‚‹å‹ã§ã‚ã£ã¦ã‚‚å¼•æ•°ã¨ã—ã¦å—ã‘å–ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚

````python
#æ–‡å­—åˆ—ã¨æ•°å€¤ã‚’å¼•æ•°ã¨ã—ã¦å—ã‘å–ã‚‹é–¢æ•°greeting
def greeting(en, number, name):
    #æ–‡å­—åˆ—ã«æ•°ã‚’æ›ã‘ç®—ã™ã‚‹ã¨ã€æ–‡å­—åˆ—ã‚’æ•°ã®å›ã ã‘ç¹°ã‚Šè¿”ã™ã“ã¨ã‚’æŒ‡å®šã—ã¾ã™
    print(en*number+','+name)

#é–¢æ•°greetingã«æ–‡å­—åˆ—ã¨æ•°å€¤ã‚’å¼•æ•°ã¨ã—ã¦æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting('Hello',3, 'World')
````

é–¢æ•°ã®å¼•æ•°ã‚„é–¢æ•°å†…ã§å®šç¾©ã•ã‚Œã‚‹å¤‰æ•°ã¯ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã®ãŸã‚ã€ãã‚Œã‚‰ã®å¤‰æ•°ã¯é–¢æ•°ã®å¤–ã‹ã‚‰ã¯å‚ç…§ã§ãã¾ã›ã‚“ã€‚

````python
#å¼•æ•°greeting_localã«æ¸¡ã•ã‚ŒãŸå€¤ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting(greeting_local):
    print(greeting_local)

greeting('Hello')

#ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ï¼ˆé–¢æ•°greetingã®å¼•æ•°ï¼‰greeting_localã‚’å‚ç…§
greeting_local
````

ä¸€æ–¹ã€å¤‰æ•°ãŒ**ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°**ã§ã‚ã‚Œã°ã€ãã‚Œã‚‰ã®å¤‰æ•°ã¯é–¢æ•°ã®å¤–ã‹ã‚‰ã‚‚ä¸­ã‹ã‚‰ã‚‚å‚ç…§ã§ãã¾ã™ã€‚
ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨ã¯ãƒ—ãƒ­ã‚°ãƒ©ãƒ å…¨ä½“ã€ã©ã“ã‹ã‚‰ã§ã‚‚åˆ©ç”¨å¯èƒ½ãªå¤‰æ•°ã§ã™ã€‚

````python
#ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°greeting_globalã®å®šç¾©
greeting_global = 'Hello'

#ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°greeting_globalã®å€¤ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting():
    print(greeting_global)

greeting()

#ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°greeting_globalã‚’å‚ç…§
greeting_global
````

ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨åŒã˜åå‰ã®å¤‰æ•°ã‚’é–¢æ•°å†…ã§å®šç¾©ã™ã‚‹ã¨ã€
ãã‚Œã¯é€šå¸¸ã¯ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨ã¯ç•°ãªã‚‹ã€é–¢æ•°å†…ã®ã¿ã§åˆ©ç”¨å¯èƒ½ãª**ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°**ã®å®šç¾©ã¨ã—ã¦æ‰±ã‚ã‚Œã¾ã™ã€‚
ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨åŒã˜åå‰ã®å¼•æ•°ã‚’ç”¨ã„ã‚‹å ´åˆã‚‚åŒæ§˜ã§ã™ã€‚

````python
#ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°greeting_globalã¨åŒã˜åå‰ã®å¤‰æ•°ã«å€¤ã‚’ä»£å…¥ã™ã‚‹é–¢æ•°greeting
def greeting():
    greeting_global = 'Bonjour'
    print(greeting_global)

greeting()

#å¤‰æ•°greeting_globalã‚’å‚ç…§
greeting_global
````

ã—ã‹ã—ã€ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨åŒåã®ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã‚’å®šç¾©ã™ã‚‹ã“ã¨ã¯ã€ä¸€èˆ¬ã«æ³¨æ„ãŒå¿…è¦ã§ã™ã€‚
ä½•æ•…ãªã‚‰ã€ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã¨ã—ã¦ã®å®šç¾©ã‚’å«ã‚€é–¢æ•°å†…ã§ã¯ã€åŒåã®ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã‚’å‚ç…§ã§ããªã„ã‹ã‚‰ã§ã™ã€‚
ãŸã¨ãˆã°ã€æ¬¡ã®ã‚³ãƒ¼ãƒ‰ã¯ã€`Hello` ã¨ `Bonjour` ãŒé †ã«å°å­—ã™ã‚‹ã“ã¨ã‚’æœŸå¾…ã™ã‚‹ã‹ã‚‚ã—ã‚Œã¾ã›ã‚“ãŒã€

````python
def greeting():
    print(greeting_global) # æœ€åˆã®å‚ç…§
    greeting_global = 'Bonjour' # ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã®å®šç¾©
    print(greeting_global)

greeting()
````

æœ€åˆã®å‚ç…§æ™‚ã«ã¯ã€ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã® `greeting_global` ãŒæœªå®šç¾©ãªã®ã§ã€ã‚¨ãƒ©ãƒ¼ãŒç”Ÿã˜ã¾ã™ã€‚

### â–² global å®£è¨€

é–¢æ•°å†…ã§ã¯ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ãŒæ›´æ–°ã•ã‚Œãªã„ã®ãŒåŸºæœ¬ã§ã™ã€‚


```
ğŸ˜€
ãã®ã¾ã¾å¤‰æ›´ã—ã‚ˆã†ã¨ã™ã‚‹ã¨ã€ãƒ­ãƒ¼ã‚«ãƒ«å¤‰æ•°ã‚’å®šç¾©ã—ãŸã¨åˆ¤å®šã•ã‚Œã¦ã—ã¾ã†ã€‚
c++ã®æ™‚ã¯ã€int x ã®ã‚ˆã†ã«å®šç¾©ã™ã‚‹ãŸã‚ã€é–¢æ•°å†…ã§ åŒåã®å¤‰æ•°ã¨ã—ã¦ x = 10ã¨ã™ã‚‹ã¨ã€
ãã®ã¾ã¾ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã¨ã—ã¦æ‰±ã£ã¦ãã‚Œã‚‹ãŒã€

```



ã—ã‹ã—ã€ã©ã†ã—ã¦ã‚‚é–¢æ•°å†…ã§ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã‚’æ›´æ–°ã—ãŸã„ã¨ãã«ã¯ã€
**`global`** å®£è¨€ã‚’ä½¿ã£ã¦æ›´æ–°ã—ãŸã„ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°ã‚’æŒ‡å®šã—ã¾ã™ã€‚

````python
#ã‚°ãƒ­ãƒ¼ãƒãƒ«å¤‰æ•°greeting_globalã«å€¤ã‚’ä»£å…¥ã™ã‚‹é–¢æ•°greeting
def greeting():
    global greeting_global
    greeting_global = 'Bonjour'
    print(greeting_global)

greeting()

##å¤‰æ•°greeting_globalã‚’å‚ç…§
greeting_global
````

## â–²ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°
ä¸Šè¨˜ã®ä¸€èˆ¬çš„ãªå¼•æ•°ï¼ˆä½ç½®å¼•æ•°ã¨ã‚‚å‘¼ã°ã‚Œã¾ã™ï¼‰ã§ã¯ã€
äº‹å‰ã«å®šç¾©ã—ãŸå¼•æ•°ã®é †ç•ªã«å¾“ã£ã¦ã€é–¢æ•°ã¯å¼•æ•°ã‚’å—ã‘å–ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ä»˜ãå¼•æ•°ï¼ˆ**ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°**ï¼‰ã‚’ä½¿ã†ã¨ã€é–¢æ•°ã¯å¼•æ•°ã®å¤‰æ•°åã¨ãã®å€¤ã®çµ„ã¿ã‚’å—ã‘å–ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚
ãã®éš›ã€å¼•æ•°ã¯é †ä¸åŒã§é–¢æ•°ã«æ¸¡ã™ã“ã¨ãŒã§ãã¾ã™ã€‚

````python
#æ–‡å­—åˆ—ã¨æ•°å€¤ã‚’å¼•æ•°ã¨ã—ã¦å—ã‘å–ã‚‹é–¢æ•°greeting
def greeting(en, number, name):
    print(en*number+','+name)

#é–¢æ•°greetingã«å¼•æ•°ã®å¤‰æ•°åã¨ãã®å€¤ã®çµ„ã¿ã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting(en='Hello', name='Japan', number=2)
````

ä½ç½®å¼•æ•°ã¨ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã‚’åˆã‚ã›ã¦ä½¿ã†å ´åˆã¯ã€æœ€åˆã«ä½ç½®å¼•æ•°ã‚’æŒ‡å®šã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

````python
#ä½ç½®å¼•æ•°ã¨ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã‚’çµ„ã¿åˆã‚ã›ãŸé–¢æ•°greetingã®å‘¼ã³å‡ºã—
greeting('Hello', name='Japan', number=2)
````

## â–²å¼•æ•°ã®åˆæœŸå€¤
é–¢æ•°ã‚’å‘¼ã³å‡ºã™éš›ã«ã€å¼•æ•°ãŒæ¸¡ã•ã‚Œãªã„å ´åˆã«ã€**åˆæœŸå€¤**ã‚’å¼•æ•°ã¨ã—ã¦æ¸¡ã™ã“ã¨ãŒã§ãã¾ã™ã€‚

åˆæœŸå€¤ã®ã‚ã‚‹å¼•æ•°ã«å€¤ã‚’æ¸¡ã—ãŸã‚‰ã€é–¢æ•°ã¯ãã®å¼•æ•°ã®åˆæœŸå€¤ã®ä»£ã‚ã‚Šã«æ¸¡ã•ã‚ŒãŸå€¤ã‚’å—ã‘å–ã‚Šã¾ã™ã€‚

åˆæœŸå€¤ã‚’æŒã¤å¼•æ•°ã¯ã€ä½ç½®å¼•æ•°ã®å¾Œã«æŒ‡å®šã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

````python
#å¼•æ•°ã®åˆæœŸå€¤ï¼ˆå¼•æ•°ã®å¤‰æ•°enã«å¯¾ã™ã‚‹'Hello'ï¼‰ã‚’æŒã¤é–¢æ•°greeting
def greeting(name, en='Hello'):
    print(en+', '+name)

#å¼•æ•°ã®åˆæœŸå€¤ã‚’æŒã¤é–¢æ•°greetingã®å‘¼ã³å‡ºã—
greeting('World')
````

## â–²å¯å¤‰é•·å¼•æ•°
ä»®å¼•æ•°ã®å‰ã« **`*`** ã‚’ä»˜ã‘ã¦é–¢æ•°ã‚’å®šç¾©ã™ã‚‹ã¨ã€è¤‡æ•°ã®å¼•æ•°ã‚’ã‚¿ãƒ—ãƒ«ã¨ã—ã¦å—ã‘å–ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚
å‘¼ã³å‡ºã™å´ã¯ã€å¼•æ•°ã®å€‹æ•°ã‚’å¤‰ãˆã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚
ä¸€èˆ¬ã«ã€å€‹æ•°ãŒå¯å¤‰ã®å¼•æ•°ã¯**å¯å¤‰é•·å¼•æ•°**ã¨å‘¼ã°ã‚Œã¾ã™ã€‚

````python
#å¯å¤‰é•·ã®å¼•æ•°ã‚’å—ã‘å–ã‚Šã€ãã‚Œã‚‰ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting(*args):
    print(args)

#å¯å¤‰é•·ã®å¼•æ•°ã‚’å—ã‘å–ã‚‹é–¢æ•°greetingã«è¤‡æ•°ã®å¼•æ•°ã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting('Hello','Bonjour','Guten Tag')
````

ãƒªã‚¹ãƒˆã‚„ã‚¿ãƒ—ãƒ«ã®è¦ç´ ã‚’å¯å¤‰é•·å¼•æ•°ã¨ã—ã¦é–¢æ•°ã«æ¸¡ã™å ´åˆã¯ã€**`*`** ã‚’ãƒªã‚¹ãƒˆã‚„ã‚¿ãƒ—ãƒ«ã®å‰ã«ã¤ã‘ã¦æ¸¡ã—ã¾ã™ã€‚

````python
#ãƒªã‚¹ãƒˆå‹ã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆgreeting_listã‚’é–¢æ•°greetingã«æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting_list = ['Hello','Bonjour','Guten Tag']
greeting(*greeting_list)
````

## â–²è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°
ä»®å¼•æ•°ã®å‰ã« <strong>`**`</strong> ã‚’ä»˜ã‘ã¦é–¢æ•°ã‚’å®šç¾©ã™ã‚‹ã¨ã€è¤‡æ•°ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã‚’è¾æ›¸ã¨ã—ã¦å—ã‘å–ã‚‹ã“ã¨ãŒã§ãã¾ã™ã€‚è¾æ›¸ã¨ã—ã¦å—ã‘å–ã‚‰ã‚Œã‚‹å¼•æ•°ã¯ã€**è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°**ã¨å‘¼ã°ã‚Œã¾ã™ã€‚

````python
#å¯å¤‰é•·ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã‚’å—ã‘å–ã‚Šã€ãã‚Œã‚‰ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting(**kwargs):
    print(kwargs)

#å¯å¤‰é•·ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã‚’å—ã‘å–ã‚‹é–¢æ•°greetingã«è¤‡æ•°ã®å¼•æ•°ã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting(en='Hello', fr='Bonjour', de='Guten Tag')
````

è¾æ›¸ã®å„ã‚­ãƒ¼ã¨å€¤ã‚’è¤‡æ•°ã®ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰å¼•æ•°ã¨ã—ã¦é–¢æ•°ã«æ¸¡ã™å ´åˆã¯ã€
`**` ã‚’ãã®è¾æ›¸ã®å‰ã«ã¤ã‘ã¦æ¸¡ã—ã¾ã™ã€‚

````python
#è¾æ›¸å‹ã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆgreeting_dictã‚’é–¢æ•°greetingã«æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting_dict = {'en': 'Hello', 'fr': 'Bonjour', 'de': 'Guten Tag'}
greeting(**greeting_dict)
````

## â–²å¼•æ•°ã®é †ç•ª
ä½ç½®å¼•æ•°ã€åˆæœŸå€¤ã‚’æŒã¤å¼•æ•°ã€å¯å¤‰é•·å¼•æ•°ã€è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°ã¯ã€åŒæ™‚ã«æŒ‡å®šã™ã‚‹ã“ã¨ãŒã§ãã¾ã™ãŒã€
ãã®éš›ã€ã“ã‚Œã‚‰ã®é †ç•ªã§æŒ‡å®šã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

```Python
def é–¢æ•°å(ä½ç½®å¼•æ•°, åˆæœŸå€¤ã‚’æŒã¤å¼•æ•°, å¯å¤‰é•·å¼•æ•°, è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°)
```

````python
#ä½ç½®å¼•æ•°ã€åˆæœŸå€¤ã‚’æŒã¤å¼•æ•°ã€å¯å¤‰é•·å¼•æ•°ã€è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°
#ãã‚Œãã‚Œã‚’å¼•æ•°ã¨ã—ã¦å—ã‘å–ã‚Šã€ãã‚Œã‚‰ã‚’è¡¨ç¤ºã™ã‚‹é–¢æ•°greeting
def greeting(greet, en='Hello', *args, **kwargs):
    print(greet)
    print(en)
    print(args)
    print(kwargs)

#å¯å¤‰é•·å¼•æ•°ã¸æ¸¡ã™ãƒªã‚¹ãƒˆ
greeting_list = ['Bonjour']

#è¾æ›¸å‹ã®å¯å¤‰é•·å¼•æ•°ã¸æ¸¡ã™è¾æ›¸
greeting_dict = {'de': 'Guten Tag'}

#é–¢æ•°greetingã«å¼•æ•°ã‚’æ¸¡ã—ã¦å‘¼ã³å‡ºã—
greeting('Hi', 'Hello', *greeting_list, **greeting_dict)
````
